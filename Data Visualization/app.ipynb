Importing Libraries

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing Dataset

```python
titanic_df = pd.read_csv('Titanic-Dataset.csv')
titanic_df
```

Information : lots of missing data in Cabin and Age

```python
titanic_df.info()
```

Visualization of missing data 

```python
sns.heatmap(titanic_df.isnull(),yticklabels=False, cbar=False, cmap='Reds')
```

Survival-rate Insights : 38.384 %

```python
titanic_df.describe()
```

Visualization of Survivals

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',data=titanic_df, hue='Survived',palette=['red','green'])
plt.xticks([0,1],['Not Survived','Survived'])
```

Visualization of Male vs Female

```python
sns.countplot(x='Sex',data=titanic_df, hue='Sex', palette=['blue','pink'])
```

Gender Based Survivals

More Females were survived as compared to Males

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',data=titanic_df, hue='Sex', palette=['blue','pink'])
plt.xticks([0,1],['Not Survived','Survived'])
```

Classes

```python
sns.countplot(x='Pclass',data=titanic_df,hue='Pclass')
```
